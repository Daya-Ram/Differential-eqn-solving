## Differential eqn solving

$\frac{dy}{dt} = -k*y$

k=0.2


y0=12

```python
import numpy as np      
from scipy.integrate import odeint   # for differential eqn solving
import matplotlib.pyplot as plt

```

```python
def model(y,t):
    k=0.2
    y0=12
    dydt=-k*y
    return dydt
    
```

```python
t=np.linspace(0,20,50)
```

```python
y=odeint(model,y0,t)
plt.plot(y,t,'r',linewidth=2)
plt.xlabel('time')
plt.title('differential eqn graph')
plt.ylabel('y(t)')
plt.grid(True)
plt.show()
```

```python
y=odeint(model,y0,t)
plt.plot(y,t,'r',linewidth=2)
plt.xlabel('time')
plt.title('differential eqn graph')
plt.ylabel('y(t)')
plt.grid(True)
plt.scatter(y,t)
plt.show()
```

### Question
$\frac{dy(t)}{dt} = -y(t) + 1$

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def model(y,t):
    dydt = -y + 1
    return dydt
y0=0
t=np.linspace(0,20,100)
y=odeint(model,y0,t)
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.show()
```

$\frac{dy}{dt} = k*y +3.0$

k = 0.8

y0 = 0.5

To solve this differential equation, let us consider a function as

```python
def model(y,t):
    dydt=k*y+3.0
    return dydt
y0=0.5 

t=np.linspace(0,20,100)
k=0.8
y=odeint(model,y0,t)
plt.plot(t,y)
plt.xlabel('time')
plt.ylabel('y(t)')
plt.grid(True)
plt.show()
```
